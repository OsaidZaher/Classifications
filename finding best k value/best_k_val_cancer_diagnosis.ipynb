# Best K value breast cancer classifier


This classifier uses k-NN to classify breast cancer data, training with k=3 and testing different k values (1-100) to find the optimal parameter. Through visualization, k=60 proved most accurate with an accuracy of over 92%.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
breast_cancer_data = load_breast_cancer()
```

```python
print(breast_cancer_data.data[0])
print(breast_cancer_data.feature_names)
print(breast_cancer_data.target)
print(breast_cancer_data.target_names)
```

```python
training_d, validation_d, training_l, validation_l = train_test_split(
    breast_cancer_data.data, breast_cancer_data.target, test_size=0.2, random_state=100)
```

```python
print(training_d.shape, training_l.shape)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(training_d, training_l)
print("Accuracy with k=3:", classifier.score(validation_d, validation_l))
```

```python
accuracies = []
k_list = list(range(1, 101))

for k in range(1, 101):
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(training_d, training_l)
    accuracy = classifier.score(validation_d, validation_l)
    accuracies.append(accuracy)
```

```python
plt.plot(k_list, accuracies)
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.title('Accuracy of k-NN Classifier for Different k Values')
plt.show()
accuracy = classifier.score(validation_d, validation_l)
print("Model Accuracy:", accuracy)
```

The best K value for the accuracy of this model is 60
